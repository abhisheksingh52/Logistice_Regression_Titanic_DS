```python
titanic_ds['Pclass'].value_counts()
```

```python
import matplotlib.pyplot as plt
def pie_chart_funct(parameters):

    titanic_ds[parameters].value_counts(sort=False).plot.pie(figsize=(6,6),autopct='%1.1f%%')
    plt.show()
    return

pie_chart_funct('Pclass')
```

```python
import matplotlib.pyplot as plt
table=pd.crosstab(titanic_ds.Pclass,titanic_ds.Survived)
table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Stacked Bar Chart of Passenger Class vs Survival rate')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.savefig('Pclass_vs_Survival')
```

#### Here we can see that survival rate is varying as per the passenger class

```python
titanic_ds['Sex'].value_counts()
```

```python
pie_chart_funct('Sex')
```

```python
table=pd.crosstab(titanic_ds.Sex,titanic_ds.Survived)
table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Stacked Bar Chart of Gender vs Survival rate')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.savefig('Gender_vs_Survival')
```

#### From the visualization,we can infer that females have more survival rate than male.

```python
titanic_ds['SibSp'].value_counts()
```

```python
def bar_chart_funct(parameters):

    titanic_ds[parameters].value_counts(sort=False).plot.bar(figsize=(15,10),legend='True')
    plt.show()
    return

bar_chart_funct('SibSp')
```

```python
pd.crosstab(titanic_ds.SibSp,titanic_ds.Survived).plot(kind='bar')
plt.title('Survival Frequency for Passengers')
plt.xlabel('# of Siblings/Spouse on ship')
plt.ylabel('Frequency of Survival')
plt.savefig('sur_fre_sib_bar')
```

#### From the visualization,we can notice that as # of siblings/spouse increases survival rate is decresing for passengers.
